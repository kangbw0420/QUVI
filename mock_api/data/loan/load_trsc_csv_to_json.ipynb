```python
import csv
import json

# 은행 코드와 은행명 매핑 데이터 (넣을까 말까?)
bank_mapping = {
    "10000002": "산업은행",
    "10000003": "기업은행",
    "10000005": "외환은행",
    "10000088": "신한은행",
    "10000011": "농협",
    # 필요한 추가 매핑을 여기에 입력
}

def transform_csv_to_json_with_bank_names(csv_file_path, json_file_path):
    # 결과를 저장할 리스트
    result = []
    
    # CSV 파일 읽기
    encodings = ['utf-8', 'cp949', 'euc-kr']
    
    for encoding in encodings:
        try:
            result = []
            with open(csv_file_path, 'r', encoding=encoding) as file:
                csv_reader = csv.DictReader(file)
                for row in csv_reader:
                    # 은행 코드로 은행명 조회
                    # bank_name = bank_mapping.get(row['bank_cd'], '알 수 없음')  # 기본값 '알 수 없음'
                    
                    # note 필드 처리 로직도 생각해봄 
                    # note_value = row.get('loan_trsc_dv', '').strip() or row.get('note1', '').strip() or row.get('note2', '').strip()

                    # 새로운 JSON 형식에 맞게 (데이터 매핑)
                    transformed_data = {
                            # "USE_INTT_ID": row["use_intt_id"],                 # 이용기관아이디
                            # "BIZ_NO": row["biz_no"],                           # 사업자번호
                            "BANK_CD": bank_mapping[row["bank_cd"]],                         # 은행코드
                            "ACCT_NO": row["acct_no"],                         # 계좌번호
                            "TRSC_DT": row["trsc_dt"],                         # 거래일자
                            "TRSC_SEQ_NO": row["trsc_seq_no"],                 # 거래일련번호
                            "CURR_CD": row["curr_cd"],                         # 통화코드
                            "LOAN_TRSC_DV": row["loan_trsc_dv"],               # 대출거래구분
                            # "ACCT_NO_EXT": row["acct_no_ext"],                 # 계좌번호_확장
                            # "TRSC_TM": row["trsc_tm"],                         # 거래시간
                            "TRSC_AMT": str(float(row["trsc_amt"])),                       # 거래금액
                            "LOAN_TRSC_AMT": str(float(row["loan_trsc_amt"])),             # 대출거래원금
                            "LOAN_BAL": str(float(row["loan_bal"])),                       # 대출잔액
                            "LOAN_RATE": str(float(row["loan_rate"])),                     # 대출이율
                            "LOAN_INTR_ST_DT": row["loan_intr_st_dt"],         # 대출이자계산시작일
                            "LOAN_INTR_EN_DT": row["loan_intr_en_dt"],         # 대출이자계산종료일
                            # "LOAN_INTR_PERIOD": row["loan_intr_period"],       # 대출이자계산기간
                            "LOAN_INTR_AMT": str(float(row["loan_intr_amt"])),             # 대출이자금액
                            # "LOAN_INTR_NRML_AMT": str(float(row["loan_intr_nrml_amt"])),   # 대출이자상세(정상이자)
                            # "LOAN_INTR_OVER_AMT": str(float(row["loan_intr_over_amt"])),   # 대출이자상세(연체이자)
                            # "LOAN_INTR_RTRN_AMT": str(float(row["loan_intr_rtrn_amt"])),   # 대출이자상세(환출이자)
                            "NOTE1": row["note1"],                             # 비고1
                            # "NOTE2": row["note2"],                             # 비고2
                            # "REG_DTM": row["reg_dtm"],                         # 등록일시
                            # "CORC_DTM": row["corc_dtm"],                       # 수정일시
                            "MNUL_WRT_YN": row["mnul_wrt_yn"],                 # 수기등록여부
                            # "CUST_TYPE_CONT": row["cust_type_cont"],           # 거래구분내용
                            # "CUST_BRN": row["cust_brn"]                        # 거래지점
        }
                    
                    result.append(transformed_data)
                with open(json_file_path, 'w', encoding='utf-8') as file:
                    json.dump({"loan_trsc": result}, file, ensure_ascii=False, indent=4)
                
                    return result
        except UnicodeDecodeError:
            continue
    # JSON 파일로 저장
    
```

```python
transform_csv_to_json_with_bank_names(csv_file_path="대출_거래내역.csv", json_file_path="loan_transaction.json")
```
