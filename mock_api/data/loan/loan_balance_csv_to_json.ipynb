```python
import csv
import json

# 은행 코드와 은행명 매핑 데이터 (넣을까 말까?)
bank_mapping = {
    "10000002": "산업은행",
    "10000003": "기업은행",
    "10000005": "외환은행",
    "10000088": "신한은행",
    "10000011": "농협",
    # 필요한 추가 매핑을 여기에 입력
}

def convert_csv_to_json(csv_file_path, json_file_path):
    # 결과를 저장할 리스트
    result = []
    
    # CSV 파일 읽기
    with open(csv_file_path, 'r', encoding='cp949') as file:
        csv_reader = csv.DictReader(file)
        
        for row in csv_reader:
            # 새로운 JSON 형식에 맞게 데이터 매핑
            transformed_data = {
                # "USE_INTT_ID": row["use_intt_id"],                 # 이용기관아이디
                # "BIZ_NO": row["biz_no"],                           # 사업자번호
                "BANK_CD": bank_mapping[row["bank_cd"]],                         # 은행코드
                "ACCT_NO": row["acct_no"],                         # 계좌번호
                "ACCT_DV": row["acct_dv"],                         # 계좌구분
                "CURR_CD": row["curr_cd"],                         # 통화코드
                # "ACCT_NO_EXT": row.get("acct_no_ext", ""),         # 계좌번호_확장
                "ACCT_NICK_NM": row["acct_nick_nm"],               # 계좌별칭
                "ACCT_BAL_AMT": str(float(row["acct_bal_amt"])),   # 계좌잔액
                "ACCT_BAL_WON": str(float(row["acct_bal_won"])),   # 계좌잔액(원화환산)
                "REAL_AMT": str(float(row["real_amt"])),           # 가용자금
                "NEW_DT": row["new_dt"],                           # 신규일자
                "DUE_DT": row["due_dt"],                           # 만기일자
                "CNTRCT_AMT": str(float(row["cntrct_amt"])),       # 약정금액
                "INTR_RATE": str(float(row["intr_rate"])),         # 이자율
                "LOAN_SBJT": row["loan_sbjt"],                     # 대출과목
                "REPAY_MTHD": row["repay_mthd"],                   # 상환방법
                # "PLDG_STTS": row["pldg_stts"],                     # 질권상태
                # "EXEC_NO": row["exec_no"],                         # 실행번호
                # "LOAN_DV": row["loan_dv"],                         # 대출구분
                # "SCRT_DTL": row["scrt_dtl"],                       # 담보내역
                "INTR_PAY_DT": row["intr_pay_dt"],                 # 이자납입예정일자
                "TRMN_YN": row["trmn_yn"],                         # 해지여부
                "TRMN_DT": row["trmn_dt"],                         # 해지일자
                # "DISP_SEQ": row["disp_seq"],                       # 출력순서
                ## "USE_YN": row["use_yn"],                           # 사용여부
                "OPEN_DT": row["open_dt"],                         # 개설일자
                ## "ACCT_BAL_UPD_DTM": row["acct_bal_upd_dtm"],       # 최종잔액변경일시
                # "ACCT_BAL_INQ_DTM": row["acct_bal_inq_dtm"],       # 최종잔액조회일시
                ## "ACCT_BAL_INQ_ERCD": row["acct_bal_inq_ercd"],     # 최종잔액조회오류코드
                ## "ACCT_BAL_INQ_ERMSG": row["acct_bal_inq_ermsg"],   # 최종잔액조회오류메시지
                ## "ACCT_HIS_UPD_DTM": row["acct_his_upd_dtm"],       # 최종거래내역변경일시
                # "ACCT_HIS_INQ_DTM": row["acct_his_inq_dtm"],       # 최종거래내역조회일시
                ## "ACCT_HIS_INQ_ERCD": row["acct_his_inq_ercd"],     # 최종거래내역조회오류코드
                ## "ACCT_HIS_INQ_ERMSG": row["acct_his_inq_ermsg"],   # 최종거래내역조회오류메시지
                "REG_DTM": row["reg_dtm"],                         # 등록일시
                "CORC_DTM": row["corc_dtm"],                       # 수정일시
                ## "REAL_CLS_YN": row["real_cls_yn"],                 # 가용시재여부
                # "EXEC_BATCH_YN": row["exec_batch_yn"],             # 일별잔액예외처리구분
                # "EXT1": row["ext1"],                               # 예비1
                # "MNUL_WRT_YN": row["mnul_wrt_yn"],                 # 수기 작성 여부
                # "CERT_REG_SEQ_NO": row["cert_reg_seq_no"],         # 인증정보일련번호
                ## "UNOTE_CERT_REG_SEQ_NO": row["unote_cert_reg_seq_no"], # 전자어음인증정보일련번호
                ## "UNOTE_YN": row["unote_yn"],                       # 전자어음계좌여부
                # "ACCT_UNOTE_INQ_DTM": row["acct_unote_inq_dtm"],   # 최종전자어음조회일시
                # "ACCT_UNOTE_INQ_ERCD": row["acct_unote_inq_ercd"], # 최종전자어음조회오류코드
                # "ACCT_UNOTE_INQ_ERMSG": row["acct_unote_inq_ermsg"], # 최종전자어음조회오류메시지
                ## "DEL_YN": row["del_yn"]                            # 삭제여부
        }
            
            result.append(transformed_data)
    
    # JSON 파일로 저장
    with open(json_file_path, 'w', encoding='utf-8') as file:
        json.dump({"loan_accounts": result}, file, ensure_ascii=False, indent=4)
    
    return result
```

```python
convert_csv_to_json(csv_file_path="대출_잔액.csv", json_file_path="loan_balance.json")
```
