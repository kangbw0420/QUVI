```python
import csv
import json

def convert_csv_to_json(csv_file_path, json_file_path):
    # 결과를 저장할 리스트
    result = []
    
    # CSV 파일 읽기
    with open(csv_file_path, 'r', encoding='utf-8') as file:
        csv_reader = csv.DictReader(file)
        
        for row in csv_reader:
            # 새로운 JSON 형식에 맞게 데이터 매핑
            transformed_data = {
                "TRSC_TM": row['거래시간'],
                "BANK_NM": row['은행명'],
                "TRSC_AMT": str(float(row['거래금액'])),  # 금액을 float으로 변환
                "TRSC_DT": row['거래일자'],
                "ACCT_NO": row['계좌번호'],
                "TRSC_BAL": str(float(row['잔액'])),  # 잔액을 float으로 변환
                "IN_OUT_DV": "2" if row['입출구분'] == '출금' else "1",  # 출금=2, 입금=1로 변환
                "NOTE1": row['적요1']
            }
            
            result.append(transformed_data)
    
    # JSON 파일로 저장
    with open(json_file_path, 'w', encoding='utf-8') as file:
        json.dump({"transactions": result}, file, ensure_ascii=False, indent=4)
    
    return result
```

```python
convert_csv_to_json('웹케시내역.csv', 'webcash_transaction.json')
```
